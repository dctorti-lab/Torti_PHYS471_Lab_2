Here we wish to re-plot the data that we extracted from the oscilloscope for our second experiment. By plotting the data for each individual experiment with an appropriate time scale, we can measure the $\Delta t$ required to calculate the speed of sound in our media.

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
filepath = "/content/drive/MyDrive/Lab_2/speed_of_sound_aluminum.csv"
data = np.genfromtxt(filepath, delimiter=',', names = ["C1","C2", "C3"])
```

```python
time = data['C1']
ch1 = data['C2']
ch2 = data['C3']
```

```python
fig, ax = plt.subplots()
ax.plot(time, ch1, color="yellow")
ax.plot(time, ch2, color="blue")
ax.set_xlabel('Time (seconds)')
ax.set_ylabel('Voltage (Volts)')
ax.set_title('Voltage vs Time for Aluminum Rod')
ax.grid
plt.show()
```

Let's plot the second graph for our copper rod:

```python
filepath = "/content/drive/MyDrive/Lab_2/speed_of_sound_copper.csv"
data = np.genfromtxt(filepath, delimiter=',', names = ["D1","D2","D3"])

time_2 = data['D1']
ch1_2 = data['D2']
ch2_2 = data['D3']
```

```python
fig, ax = plt.subplots()
ax.plot(time_2, ch1_2, color="yellow")
ax.plot(time_2, ch2_2, color="blue")
ax.set_xlabel('Time (seconds)')
ax.set_ylabel('Voltage (Volts)')
ax.set_title('Voltage vs Time for Copper Rod')
ax.grid
plt.show()
```
